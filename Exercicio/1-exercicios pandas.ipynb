### 1. Ler e exibir um arquivo CSV usando pandas

```
import numpy as np
import pandas as pd

df = spark.read.csv("/FileStore/tables/Retail-1.csv", header=True, sep=',')
df = df.toPandas()

df.head(10)
```

### 2. Selecionar colunas específicas de um DataFrame usando pandas

```
colunas_selecionadas = ['region', 'description']
df_selecionado = df[colunas_selecionadas]

df_selecionado.head()
```

### 3. Filtrar linhas com base em uma condição usando pandas

```
df.loc[df['region'] == 'United Kingdom']
```

### 4. Agrupar dados com base em uma coluna usando pandas

```
df_agrupado = df.groupby("invoice_no").agg({'quantity': 'sum'}).reset_index()

print(df_agrupado)
```

### 5. Usar a função max em alguma coluna do DataFrame

```
max_unit_price = df['unit_price'].max()

print(max_unit_price)
```
